```python
def avg(review):
    words = review.split()
    return (sum(len(word) for word in words) / len(words))
```

```python
df['avg_word'] = df['reviews'].apply(lambda x: avg(x))
```

```python
df['reviews_lower'] = df['reviews'].apply(lambda x: x.lower())
```

```python
df.head()
```

```python
import string
import re

table = str.maketrans('', '', string.punctuation+string.digits)
```

```python
df['rmv_punc'] = df["reviews_lower"].apply(lambda x: re.sub(' +', ' ',x.translate(table)))
```

```python
df['rmv_punc_stp'] = df['rmv_punc'].apply(lambda x: " ".join([r for r in x.split() if r not in stopwords]))
```

```python
df['rmv_punc_stp'] = df["rmv_punc_stp"].apply(lambda x: re.sub(' +', ' ',x))
```

```python
freq= pd.Series(" ".join(df['rmv_punc_stp']).split()).value_counts()[:30]
freq
```
